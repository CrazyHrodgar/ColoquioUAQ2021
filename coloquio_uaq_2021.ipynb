Un año es bisiesto si es cumple con estos criterios
	- Es bisiesto si es divisible entre 4
	- No es bisiesto si es divisible entre 100
	- Es bisiesto si es divisible entre 400
	
	Probar con los siguientes valores
	1900 --> False
	2000 --> True
	1920 --> True
	2044 --> True
	1993 --> False
	1990 --> False

```python
# Programa que diga si un año es bisiesto o no

def leap_year(year):
	"""
	Un año es bisiesto si es cumple con estos criterios
	- Es bisiesto si es divisible entre 4
	- No es bisiesto si es divisible entre 100
	- Es bisiesto si es divisible entre 400
	
	Probar con los siguientes valores
	1900 --> False
	2000 --> True
	1920 --> True
	2044 --> True
	1993 --> False
	1990 --> False
	"""
	if (year % 4) == 0: # Primera condición
		if (year % 100) == 0: # Segunda condición
			if (year % 400) == 0: # Tercera condición
				# return True
				print('Año bisiesto')
			else:
				# return False
				print('Año normal')
		else:
			# return True
			print('Año bisiesto')
	else:
		# return False
		print('Año normal')
```

El programa devuelve el componente inverso de una cadena de ADN

Si el usuario ingresa una A devuelve una C, si el usuario ingresa una G devuelve una T y visceversa.

El dato se debe pasar como una string, no como una lista de caracteres

Ejemplo
ACTGCA --> CAGTAC

```python
def dna(chain):
	"""
	El programa devuelve el componente inverso de una cadena de ADN
	Si el usuario ingresa una A devuelve una C, si el usuario ingresa una G devuelve una T y visceversa.
	El dato se debe pasar como una string, no como una lista de caracteres
	Ejemplo
	ACTGCA --> CAGTAC
	"""
	# La solución sencilla es con un ciclo for que itere a través de la string y que devuelva el valor inverso
	# La solución más elegante es hacer un diccionario con los resultados y que se devuelva el valor correspondiente

	adn = {'A':'C',
	'G':'T',
	'C':'A',
	'T':'G'
	}

	for i in chain:
		print(adn[i])
```

```python
dna('ATCGCCATG')
```

```python
def volumen(radio,condition, h=25):
	"""
	La variable radio hace referencia al radio del circulo
	La variable condition será Booleana y servirá para indicar
	La variable h será la altura del cílindro, dejamos un valor por defecto de 25.
	al programa si queremos obtener el volumen de la esfera
	"""
	pi = 3.1416 # El valor de pi

	area = pi*radio**2

	if condition == True:
		vol = area*h
	else:
		vol = 'No se solicitó el volumen'

	return(area,vol,'Terminado')
```
